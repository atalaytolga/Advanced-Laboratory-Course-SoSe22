## 00_Au_bars

```python
filenames = glob.glob('./data_sample_5F/00_Au_bars/*.txt')
```

```python
for f in filenames:
    print(f)
    df = pd.read_csv(f, sep="	",header=None, skiprows=3)
    x=df.to_numpy()[0]
    y=df.to_numpy()[1]
    x*=10**6
    y*=10**6
    plt.plot(x,y)
    plt.show()
```

```python
from scipy import stats
slope, intercept, r, p, se = stats.linregress(x, y)
```

```python
se
```

```python
slope
```

```python
xn=np.linspace(-1,1,100)
plt.scatter(x,y,label="Experimental Data")
plt.plot(xn,xn*slope+intercept,c="#00B945",label="Linear Fit")


plt.annotate(r'$R=6.80 \pm 0.15 \Omega$', xy=(0.3, 23),  xycoords='data',backgroundcolor="w"
            )

plt.title(r"IV characteristic of gold sample")

plt.xlabel(r"Source current[mA]")
plt.ylabel(r"Voltage[mV]")

plt.legend(scatterpoints=3)
plt.show()
```
